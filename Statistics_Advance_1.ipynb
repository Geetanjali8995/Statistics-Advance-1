Q.13.  What is the Central Limit Theorem (CLT), and why is it critical in statistics?

Ans:The Central Limit Theorem (CLT) is one of the most important concepts in statistics.

**Definition:**
The Central Limit Theorem states that:

"Regardless of the population distribution shape, the sampling distribution of the sample mean approaches a normal distribution as the sample size becomes large."

Formally:
If you take random samples of size n from any population with mean μ and standard deviation σ, then the sampling distribution of the sample mean (𝑋̄) will be approximately normal if n is sufficiently large (typically n ≥ 30).

**Why CLT Is Critical in Statistics:**

1. Enables Use of Normal Distribution:

  Even if the data is not normally distributed, CLT lets us use normal probability models for inference with large enough samples.

2. Foundation of Inferential Statistics:

  Forms the basis for confidence intervals and hypothesis testing for means and proportions.

3. Simplifies Analysis:

  Allows us to apply Z-tests and t-tests to real-world data that may not be perfectly normal.

4. Used Across Fields:

  Widely applied in business, economics, psychology, data science, and quality control.

Q.14.  How does the Central Limit Theorem relate to the normal distribution?

Ans: **Relationship Between CLT and the Normal Distribution:**

**1.CLT Explains Why the Normal Distribution Is So Common**

Even if the underlying population distribution is not normal, the distribution of sample means (or sums) becomes approximately normal as the sample size increases.

**2.Normal Distribution Emerges from CLT**

The CLT guarantees that the sampling distribution of the sample mean approaches a normal distribution with:

- Mean: μ (same as population mean)

- Standard deviation

**3.Normal Approximation in Practice**

Thanks to the CLT, we can use the standard normal distribution (Z-distribution) to:

- Calculate probabilities

- Construct confidence intervals

- Perform hypothesis tests
  even if the original data is not normally distributed—as long as the sample size is large enough.

**4.CLT Justifies Statistical Methods Based on the Normal Distribution**

Many statistical procedures (like Z-tests, t-tests, ANOVA) assume normality. The CLT justifies applying these methods when working with sample means or proportions from large samples.

Q.18.  What is the significance of confidence intervals in statistical analysis?

Ans:

**Significance of Confidence Intervals in Statistical Analysis**

A confidence interval (CI) provides a range of values that likely contains the true population parameter (like the mean or proportion), based on sample data and a chosen confidence level.

**Why Confidence Intervals Are Important:**

1. Quantify Uncertainty
  Unlike a single point estimate, CIs reflect the uncertainty due to sampling variability.

  They help answer:
 "How precise is our estimate?"

2. Provide a Range for Decision Making

  CIs give a plausible range where the true parameter may lie.

  For example: “We are 95% confident the true mean salary is between ₹50,000 and ₹55,000.”

3. Enable Comparison Between Groups

  If two groups' confidence intervals overlap significantly, we may conclude there's no strong difference.

  If they don't overlap, it often suggests a statistically significant difference.

4. Used in Hypothesis Testing

  A confidence interval that excludes the null value (e.g., 0 for mean difference, or 0.5 for proportions) often leads to rejection of the null hypothesis.

  Example: If a 95% CI for mean difference is [2.1, 5.3], we can say the difference is statistically significant.

5. Interpretable and Actionable

  Easier to interpret than p-values for non-technical stakeholders.

  Helps in making informed decisions in fields like medicine, business, and public policy.

Q.21.  What are the assumptions for applying the Central Limit Theorem?

Ans:
**Assumptions for Applying the Central Limit Theorem (CLT)**

The Central Limit Theorem (CLT) allows us to assume that the sampling distribution of the sample mean is approximately normal, but certain conditions must be met.

1. Random Sampling
    - The sample must be drawn randomly from the population.

    - Ensures that every member has an equal chance of selection.

👉 This reduces bias and supports representativeness.

2. Independence of Observations

    - Each observation in the sample must be independent of the others.

    - For sampling without replacement, the sample size should not exceed 10% of the population (i.e., n ≤ 0.10 × N) to maintain approximate independence.

3. Sample Size (n) Should Be "Large Enough"

    - If population is normal → CLT applies for any sample size.

    - If population is not normal → A larger sample is required (commonly n ≥ 30).

    - The more skewed the population, the larger the sample needed for the sampling distribution to approximate normality.

4. Finite Variance
    - The population must have a finite standard deviation (σ).

    - If the population variance is infinite or undefined, CLT doesn't hold.

```python
#3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
def bernoulli_pmf(x, p):
    """
    Calculate the PMF of a Bernoulli distribution.

    Parameters:
    x (int): The outcome, should be 0 or 1
    p (float): Probability of success (must be between 0 and 1)

    Returns:
    float: Probability of observing outcome x
    """
    if x not in [0, 1]:
        raise ValueError("Bernoulli outcomes must be 0 or 1.")
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1.")

    return p if x == 1 else (1 - p)
print("P(X=1):", bernoulli_pmf(1, 0.7))  # Should return 0.7
print("P(X=0):", bernoulli_pmf(0, 0.7))  # Should return 0.3

```

```python
#6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
#uniform distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Parameters for the discrete uniform distribution (e.g., die roll)
low = 1   # inclusive
high = 7  # exclusive in scipy's randint → actual values: 1 to 6

# Create the distribution
dist = randint(low, high)

# Define the range of values
x = np.arange(low, high)

# Calculate the CDF values
cdf_values = dist.cdf(x)

# Plot the CDF
plt.figure(figsize=(8, 5))
plt.step(x, cdf_values, where='post', label='CDF', color='blue')
plt.title("CDF of Discrete Uniform Distribution (1–6)")
plt.xlabel("Outcome")
plt.ylabel("Cumulative Probability")
plt.xticks(x)
plt.ylim(0, 1.1)
plt.grid(True)
plt.legend()
plt.show()
```

```python
#9.  Write a Python function to calculate Z-scores from a dataset and plot them
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    """
    Calculate Z-scores from a dataset and plot them.

    Parameters:
    data (list or numpy array): Input numeric dataset
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)

    if std == 0:
        raise ValueError("Standard deviation is zero. Z-scores cannot be calculated.")

    z_scores = (data - mean) / std

    # Plot Z-scores
    plt.figure(figsize=(8, 5))
    plt.scatter(range(len(z_scores)), z_scores, color='blue', label='Z-scores')
    plt.axhline(0, color='gray', linestyle='--')
    plt.axhline(2, color='red', linestyle='--', label='Z = ±2')
    plt.axhline(-2, color='red', linestyle='--')
    plt.title('Z-Scores of the Dataset')
    plt.xlabel('Data Point Index')
    plt.ylabel('Z-Score')
    plt.legend()
    plt.grid(True)
    plt.show()

    return z_scores
    # Sample data
sample_data = [55, 60, 65, 70, 75, 80, 85, 90, 95, 100]

# Call the function
z = calculate_and_plot_zscores(sample_data)
print("Z-scores:", z)

```

```python
#10.  Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
import numpy as np
import matplotlib.pyplot as plt

def simulate_clt(population_dist, dist_params, sample_sizes, num_samples=1000):
    plt.figure(figsize=(15, 4 * len(sample_sizes)))

    for i, n in enumerate(sample_sizes):
        sample_means = []

        for _ in range(num_samples):
            sample = population_dist(*dist_params, size=n)
            sample_means.append(np.mean(sample))

        # Plot histogram of sample means
        plt.subplot(len(sample_sizes), 1, i + 1)
        plt.hist(sample_means, bins=30, density=True, color='skyblue', edgecolor='black')
        plt.title(f'CLT Approximation (n={n}, {num_samples} samples)')
        plt.xlabel('Sample Mean')
        plt.ylabel('Density')
        plt.grid(True)

    plt.tight_layout()
    plt.show()

# Simulate CLT using exponential distribution (non-normal)
simulate_clt(np.random.exponential, dist_params=(1.0,), sample_sizes=[5, 30, 100])
```

```python
#14.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
#distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    if std == 0:
        raise ValueError("Standard deviation cannot be zero.")
    return (x - mean) / std

def plot_z_score(z_score):
    # X range for standard normal distribution
    x_vals = np.linspace(-4, 4, 1000)
    y_vals = norm.pdf(x_vals, 0, 1)

    # Plot standard normal distribution
    plt.figure(figsize=(10, 5))
    plt.plot(x_vals, y_vals, label='Standard Normal Distribution', color='blue')

    # Shade the area under the curve up to the Z-score
    plt.fill_between(x_vals, y_vals, where=(x_vals <= z_score), color='lightblue', alpha=0.5, label=f'P(Z ≤ {z_score:.2f})')

    # Draw Z-score line
    plt.axvline(z_score, color='red', linestyle='--', label=f'Z = {z_score:.2f}')

    # Plot formatting
    plt.title("Z-score in Standard Normal Distribution")
    plt.xlabel("Z")
    plt.ylabel("Probability Density")
    plt.grid(True)
    plt.legend()
    plt.show()

# Example input
x = 75           # Data point
mean = 70        # Mean of the population
std = 5          # Standard deviation of the population

# Calculate and plot Z-score
z = calculate_z_score(x, mean, std)
print(f"Z-score for x = {x}: {z:.2f}")

plot_z_score(z)
```

```python
#Q.15  Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
from scipy.stats import norm

def z_test(sample, mu0, sigma, alpha=0.05, alternative='two-sided'):
    """
    Perform a one-sample Z-test.

    Parameters:
    - sample       : array-like of observed data
    - mu0          : hypothesized population mean under H0
    - sigma        : known population standard deviation
    - alpha        : significance level (default 0.05)
    - alternative  : 'two-sided', 'greater' (mu > mu0), or 'less' (mu < mu0)

    Returns:
    - z_stat       : computed Z statistic
    - p_value      : corresponding p-value
    - reject_H0    : True if H0 is rejected at level alpha
    """
    n = len(sample)
    x_bar = np.mean(sample)
    se = sigma / np.sqrt(n)
    z_stat = (x_bar - mu0) / se

    # Compute p-value
    if alternative == 'two-sided':
        p_value = 2 * (1 - norm.cdf(abs(z_stat)))
    elif alternative == 'greater':
        p_value = 1 - norm.cdf(z_stat)
    else:  # 'less'
        p_value = norm.cdf(z_stat)

    reject_H0 = p_value < alpha
    return z_stat, p_value, reject_H0

# --- Example usage ---

# Sample dataset (e.g., measured weights)
sample_data = np.array([52.1, 49.8, 51.3, 50.5, 48.9, 52.7, 51.0, 50.2, 49.5, 51.8])

# Known population standard deviation
sigma_pop = 2.0

# Hypothesized mean
mu_null = 50.0

# Significance level
alpha = 0.05

# Perform two-sided Z-test
z_stat, p_val, reject = z_test(sample_data, mu0=mu_null, sigma=sigma_pop,
                               alpha=alpha, alternative='two-sided')

print(f"Sample mean      = {sample_data.mean():.3f}")
print(f"Z-statistic      = {z_stat:.3f}")
print(f"P-value (2‑tailed)= {p_val:.4f}")
print(f"Reject H0?        = {'Yes' if reject else 'No'} (α = {alpha})")
```

```python
#Q.16  Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
from scipy.stats import norm

def confidence_interval(data, sigma, confidence=0.95):
    """
    Calculate the confidence interval for the mean using Z-distribution.

    Parameters:
    - data: array-like dataset
    - sigma: known population standard deviation
    - confidence: confidence level (default is 0.95)

    Returns:
    - (lower_bound, upper_bound): the confidence interval
    """
    n = len(data)
    mean = np.mean(data)
    z = norm.ppf((1 + confidence) / 2)
    margin_error = z * (sigma / np.sqrt(n))
    lower = mean - margin_error
    upper = mean + margin_error
    return (lower, upper), mean, margin_error

# --- Example usage ---

# Sample dataset (e.g., weights, exam scores)
data = np.array([52.1, 49.8, 51.3, 50.5, 48.9, 52.7, 51.0, 50.2, 49.5, 51.8])

# Known population standard deviation
sigma = 2.0

# Confidence level
confidence = 0.95

# Calculate confidence interval
(ci_lower, ci_upper), mean, moe = confidence_interval(data, sigma, confidence)

# Print results
print(f"Sample mean = {mean:.2f}")
print(f"{int(confidence*100)}% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
print(f"Margin of Error = ±{moe:.2f}")
```

```python
#Q.17  Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
import numpy as np
from scipy import stats

# Step 1: Generate normal data
np.random.seed(42)  # For reproducibility
mu_true = 70        # True population mean (for simulation)
sigma_true = 10     # True population std dev
n = 30              # Sample size

data = np.random.normal(loc=mu_true, scale=sigma_true, size=n)

# Step 2: Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample std deviation

# Step 3: Calculate 95% confidence interval using t-distribution
confidence = 0.95
alpha = 1 - confidence
df = n - 1
t_critical = stats.t.ppf(1 - alpha/2, df)
margin_error = t_critical * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_error
ci_upper = sample_mean + margin_error

# Step 4: Print the results
print(f"Sample Mean           = {sample_mean:.2f}")
print(f"Sample Std Deviation  = {sample_std:.2f}")
print(f"{int(confidence*100)}% Confidence Interval = ({ci_lower:.2f}, {ci_upper:.2f})")
print(f"Margin of Error       = ±{margin_error:.2f}")
```

```python
#Q.23  Implement Z-score normalization on a dataset and visualize its transformation
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate original dataset (normal but not standardized)
np.random.seed(42)
original_data = np.random.normal(loc=100, scale=20, size=1000)  # mean=100, std=20

# Step 2: Perform Z-score normalization
mean = np.mean(original_data)
std = np.std(original_data)
z_normalized_data = (original_data - mean) / std

# Step 3: Visualization - Before and After
plt.figure(figsize=(12, 5))

# Original data histogram
plt.subplot(1, 2, 1)
plt.hist(original_data, bins=30, color='skyblue', edgecolor='black')
plt.title('Original Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.axvline(mean, color='red', linestyle='--', label=f'Mean = {mean:.2f}')
plt.legend()

# Z-normalized data histogram
plt.subplot(1, 2, 2)
plt.hist(z_normalized_data, bins=30, color='lightgreen', edgecolor='black')
plt.title('Z-Score Normalized Data')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.axvline(0, color='red', linestyle='--', label='Mean = 0')
plt.legend()

plt.tight_layout()
plt.show()

# Step 4: Print summary
print(f"Original Mean: {mean:.2f}, Std Dev: {std:.2f}")
print(f"Normalized Mean: {np.mean(z_normalized_data):.2f}, Std Dev: {np.std(z_normalized_data):.2f}")
```

```python
#Q.24.  Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
#distribution
import numpy as np
from scipy.stats import skew, kurtosis

def analyze_distribution_skew_kurtosis(mean=0, std=1, size=1000):
    """
    Generates normal data and calculates skewness and kurtosis.

    Parameters:
    - mean: Mean of the normal distribution
    - std: Standard deviation
    - size: Number of data points

    Returns:
    - skew_val: Measure of symmetry (≈ 0 for normal)
    - kurt_val: Measure of tail weight (≈ 3 for normal with Fisher=False)
    """
    # Generate normal data
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Calculate skewness and kurtosis
    skew_val = skew(data)
    kurt_val = kurtosis(data, fisher=False)  # Set fisher=True to get "excess kurtosis"

    print(f"Skewness      : {skew_val:.4f} (0 indicates symmetric)")
    print(f"Kurtosis      : {kurt_val:.4f} (≈3 indicates normal shape)")

    return skew_val, kurt_val
analyze_distribution_skew_kurtosis(mean=0, std=1, size=1000)
```
